# TASK 4

Performs some normalisation techniques such as, min-max, log scaling on the GDP data (1960-2019)
of all the five countries.


```python
africa_gdp_3 = africa_gdp.copy()
```

```python
africa_gdp_3["1988"] = africa_gdp_3["1988"]/africa_gdp_3["1988"].max()

africa_gdp_3.head()
```
